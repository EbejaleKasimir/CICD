```python
import pandas as pd

left = pd.DataFrame({
    'key1': ['K0', 'K0', 'K1', 'K2'],
    'key2': ['K0', 'K1', 'K0', 'K1'],
    'A': ['AB', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

right = pd.DataFrame({
    'key1': ['K0', 'K1', 'K1', 'K2'],
    'key2': ['K0', 'K0', 'K0', 'K0'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})
```

```python
left
```

```python
right
```

Row 0: It matches K0 and K0 in both DataFrames. The corresponding values for columns A, B, C, and D are filled with the respective values from the right DataFrame.

Row 1: It matches K0 and K1 in the left DataFrame, but there is no matching row in the right DataFrame. Hence, the values for columns C and D are filled with NaN.

Row 2: It matches K1 and K0 in both DataFrames. The corresponding values for columns A, B, C, and D are filled with the respective values from the right DataFrame.

Row 3: It matches K1 and K0 in both DataFrames. Since there are multiple matches, all the matching rows from the right DataFrame are included in the result.

Row 4: It matches K2 and K1 in the left DataFrame, but there is no matching row in the right DataFrame. Hence, the values for columns C and D are filled with NaN.

In summary, the pd.merge operation combines the data from the left and right DataFrames based on the specified key columns and the specified merge type (left in this case).

```python
pd.merge(left, right, how = 'left', on = ['key1', 'key2'])
```

# Multiindexing
The multindexing docs.

Creating a MultiIndex from a labeled frame

```python
df = pd.DataFrame(
    {
        "row": [0, 1, 2],
        "One_X": [1.1, 1.1, 1.1],
        "One_Y": [1.2, 1.2, 1.2],
        "Two_X": [1.11, 1.11, 1.11],
        "Two_Y": [1.22, 1.22, 1.22],
    }
)


df
```

```python
# As Labelled Index
df = df.set_index("row")

df
```

```python

# With Hierarchical Columns
df.columns = pd.MultiIndex.from_tuples([tuple(c.split("_")) for c in df.columns])

df
```

```python
# Now stack & Reset
df = df.stack(0).reset_index(1)

df
```

```python
# And fix the labels (Notice the label 'level_1' got added automatically)
df.columns = ["Sample", "All_X", "All_Y"]

df
```
