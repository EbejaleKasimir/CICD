```python
import psycopg2
import csv
import requests

# connect to the prosgreSQL database

conn = psycopg2.connect(
    host = "localhost",
    database = "cohort_4_",
    user = "postgres",
    password = "P@55w076"
)


# Open a cursor to perform database operations
cur = conn.cursor()

# Create the table
cur.execute("""
-- Table: public.apple_products

-- DROP TABLE IF EXISTS public.apple_products;

CREATE TABLE IF NOT EXISTS public.apple_products
    (
        product_name character varying COLLATE pg_catalog."default",
        product_url character varying COLLATE pg_catalog."default",
        brand character varying COLLATE pg_catalog."default",
        sale_price integer,
        mrp integer,
        discount_percentage integer,
        number_of_ratings integer,
        number_of_reviews integer,
        upc character varying COLLATE pg_catalog."default",
        star_rating double precision,
        ram character varying COLLATE pg_catalog."default"
    )

    TABLESPACE pg_default;

    ALTER TABLE IF EXISTS public.apple_products
        OWNER to postgres;
""")


# Read data from the csv file and insert it into the table

with open('apple_products.csv') as f:
        reader = csv.reader(f)
        next (reader)  #skip header row

        for row in reader:
            if len(row) == 11: # check for the correct number of columns

                # replace empty values with default values
                row = [value if value != '' else '0' for value in row]
            cur.execute("""
                INSERT INTO apple_products (product_name, product_url, brand, sale_price, mrp, discount_percentage, 
                number_of_ratings, number_of_reviews, upc, star_rating, ram)
                VALUES(%s, %s, %s, %s, %s, %s,%s, %s, %s, %s, %s);
            """, row)
    
# commit the changes to database

conn.commit()

# close the cursor and connection

cur.close()
conn.close()
```
